```python
# ===============================
# 1. Import Libraries
# ===============================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ===============================
# 2. Load Dataset
# ===============================

df = pd.read_csv("C:\\Users\\SAYAN DAS\\OneDrive\\Desktop\\TASK_21_SAYAN\\penguins.csv")

print("First 5 rows of dataset:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

# ===============================
# 3. Data Cleaning
# ===============================

print("\nMissing Values:")
print(df.isnull().sum())

# Remove missing values
df = df.dropna()

# Remove duplicates
df = df.drop_duplicates()

print("\nDataset Shape After Cleaning:", df.shape)

# ===============================
# 4. Basic KPIs
# ===============================

total_penguins = df.shape[0]
avg_body_mass = df['body_mass_g'].mean()
avg_flipper = df['flipper_length_mm'].mean()
avg_bill_length = df['bill_length_mm'].mean()

print("\n===== KPI RESULTS =====")
print("Total Penguins:", total_penguins)
print("Average Body Mass:", round(avg_body_mass,2))
print("Average Flipper Length:", round(avg_flipper,2))
print("Average Bill Length:", round(avg_bill_length,2))

# ===============================
# 5. Species Distribution
# ===============================

species_count = df['species'].value_counts()

plt.figure()
species_count.plot(kind='bar')
plt.title("Species Distribution")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()

# ===============================
# 6. Island Distribution
# ===============================

plt.figure()
df['island'].value_counts().plot(kind='bar')
plt.title("Island Distribution")
plt.xlabel("Island")
plt.ylabel("Count")
plt.show()

# ===============================
# 7. Gender Distribution
# ===============================

plt.figure()
df['sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()

# ===============================
# 8. Average Body Mass by Species
# ===============================

species_mass = df.groupby('species')['body_mass_g'].mean()

plt.figure()
species_mass.plot(kind='bar')
plt.title("Average Body Mass by Species")
plt.xlabel("Species")
plt.ylabel("Body Mass")
plt.show()

# ===============================
# 9. Histogram - Body Mass
# ===============================

plt.figure()
df['body_mass_g'].plot(kind='hist', bins=20)
plt.title("Body Mass Distribution")
plt.xlabel("Body Mass")
plt.show()

# ===============================
# 10. Scatter Plot
# ===============================

plt.figure()
sns.scatterplot(data=df, x='flipper_length_mm', y='body_mass_g', hue='species')
plt.title("Flipper Length vs Body Mass")
plt.show()

# ===============================
# 11. Correlation Analysis
# ===============================

corr = df.corr(numeric_only=True)

print("\nCorrelation Matrix:")
print(corr)

plt.figure()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()

# ===============================
# 12. Island vs Species (Stacked Chart)
# ===============================

island_species = pd.crosstab(df['island'], df['species'])

plt.figure()
island_species.plot(kind='bar', stacked=True)
plt.title("Species Distribution by Island")
plt.xlabel("Island")
plt.ylabel("Count")
plt.show()

# ===============================
# 13. Insights
# ===============================

print("\n===== INSIGHTS =====")
print("1. Dataset contains", total_penguins, "penguins.")
print("2. Body mass and flipper length are positively correlated.")
print("3. Gentoo penguins generally have higher body mass.")
print("4. Species distribution varies across islands.")
```
